```python
y=data.quality
x=data.drop('quality',axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# Scaling the data to tkae account of variations in mean and standard deviations

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
def models(x_train,y_train):
    from sklearn.linear_model import LogisticRegression
    log=LogisticRegression(random_state=0)
    log.fit(x_train,y_train)
    #using logistic regression algorithm to the training set
    
    from sklearn.neighbors import KNeighborsClassifier
    knn=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
    knn.fit(x_train,y_train)
    #using kneighboursclassifier method of neighbours class to use nearest neighbor algorithm
    
    from sklearn.svm import SVC 
    svc_lin=SVC(kernel='linear',random_state=0)
    svc_lin.fit(x_train,y_train)
    #using SCV method of svm class to use support vector machine algorithm

    from sklearn.svm import SVC 
    svc_rbf=SVC(kernel='rbf',random_state=0)
    svc_rbf.fit(x_train,y_train)
    #using SCV method of svm class to use kernel SVM algorithm
    
    from sklearn.naive_bayes import GaussianNB
    gauss=GaussianNB()
    gauss.fit(x_train,y_train)
    #using gaussianNB method of navies_byes to use naive bayes algorithm
    
    
    from sklearn.tree import DecisionTreeClassifier
    tree=DecisionTreeClassifier(criterion='entropy',random_state=0)
    tree.fit(x_train,y_train)
    #using decisiontressclassifier method of tree class to use naive bayes algorithm
    
    from sklearn.ensemble import RandomForestClassifier
    forest=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
    forest.fit(x_train,y_train)
    #using randomforestclassifier method of ensemble class to use random forest classification algorithm
    
    
    #print model accurancy on the training data
    print('[0]Logistic Regression Training Accurancy:',log.score(x_train,y_train))
    print('[1]K Nearest Neigbor Training Accurancy:',knn.score(x_train,y_train))
    print('[2]Support vector Machine (linear classifier) Training Accurancy:',svc_lin.score(x_train,y_train))
    print('[3]Support vector machine (RBF classifier) Training Accurancy:',svc_rbf.score(x_train,y_train))
    print('[4]Gaussian Navie Bayes Training Accurancy:',gauss.score(x_train,y_train))
    print('[5]Decision Tree Classifier Training Accurancy:',tree.score(x_train,y_train))
    print('[6]Random Forest Classifier Training Accurancy:',forest.score(x_train,y_train))
    
    
    return log,knn,svc_lin,svc_rbf,gauss,tree,forest
    

    
    
```

# Evaluating Performance on Training sets

```python
model=models(x_train,y_train)
```

# Evaluating Performance on Testing datasets with Confusion Matrices

```python
from sklearn.metrics import confusion_matrix
for i in range(len(model)):
    cm=confusion_matrix(y_test,model[i].predict(x_test))
    #extracting TN,FP,FN,TP
    TN,FP,FN,TP=confusion_matrix(y_test,model[i].predict(x_test)).ravel()
    print(cm)
    print('model[()] Testing Accurancy ="{} !"'.format(i,(TP+TN)/(TP+TN+FN+FP)))
    print()
```
